# Исследование данных о продаже компьютерных игр

Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation)

**Задачи исследования:**\
Выявить определяющие успешность игры закономерности.

**Структура данных:**\
Name — название игры\
Platform — платформа\
Year_of_Release — год выпуска\
Genre — жанр игры\
NA_sales — продажи в Северной Америке (миллионы проданных копий)\
EU_sales — продажи в Европе (миллионы проданных копий)\
JP_sales — продажи в Японии (миллионы проданных копий)\
Other_sales — продажи в других странах (миллионы проданных копий)\
Critic_Score — оценка критиков (максимум 100)\
User_Score — оценка пользователей (максимум 10)\
Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

**Ход исследования:**\
Данные хранятся в файле games.csv.\
О качестве данных информации нет. Перед проведением исследования понадобится обзор данных.

1. Обзор данных, поиск ошибок, предобработка;\

5. Общий вывод.

### Проверка типов данных

В результате обзора данных в п. 1.1 можно сделать вывод о необходимости преобразования типов данных в стобцах 'year_of_release' и 'user_score'.\

## Исследовательский анализ данных

### Кол-во игр, выпускавшиеся в разные годы

```python
#построим сводную таблицу кол-ва игр по годам:
game_count = df.pivot_table(index='year_of_release', values='name', aggfunc='count')
game_count.columns = ['count']
game_count
```

```python
#построим сводную таблицу суммы продаж игр по годам:
game_sum = df.pivot_table(index='year_of_release', values='total_sales', aggfunc='sum')
game_sum.columns = ['sum']
game_sum
```

```python
#построим график кол-ва выпускаемых игр и суммы их продаж по годам и посмотрим как они соотносятся:
ax = game_count.plot(figsize=(20,8), grid = True, title = 'Кол-во выпускаемых игр по годам и сумма их продаж', label='count')
game_sum.plot(figsize=(20,8), grid = True, ax=ax, label='sum')
plt.show
```

**Выводы по полученным данным:**
1. С 1980 до 2001 года наблюдается относительно плавный рост выпуска игр и их продаж;
2. Начиная с 2002 года кол-во выпускаемых игр значительно увеличивается, при этом продажи продолжают расти более плавно;
3. Максимальное кол-во выпуска наблюдается в -2008-2009 году, после чего наблюдается спад. А максимальные продажи наблюдаются в 2008 году, после в 2009 небольшой спад, а в 2010 спад становится более существенным;
4. Начиная с 2012 года спад выпуска и продаж становится более значительным по сравнению с предыдущими годами;
5. Для оценки продаж в следующем 2017 году считаю целесообразно взять данные за последние 5 лет, т.е. с 2012 по 2016 годы, как  наиболее близкие по своим показателям. 

### Продажи по платформам

```python
#составим сводную таблицу продаж по платформам:
platform_sum = df.pivot_table(index='platform', values='total_sales', aggfunc='sum')
platform_sum.columns = ['sum']
#отсортируем по увеличению продаж и выведем первые 10 строк:
platform_sum.sort_values(by='sum', ascending=False).head(10)
```

Построим распределение по годам для каждой платформы из топ-10:

```python
platforms_max = ['PS2', 'X360', 'PS3', 'Wii', 'DS', 'PS', 'GBA', 'PS4', 'PSP', 'PC']
for i in platforms_max:
    df_i = df[df['platform'] == i]
    df_i.plot(x='year_of_release', y='total_sales', kind='scatter', ylim=(0,10), grid=True, alpha=0.5, label=i)
    plt.show()
```

**Выводы по полученным данным:**\
Стандартный срок "жизни" платформы составляет около 10 лет.

```python
df_uscore = df[~df['user_score'].isna()]
```

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Сформулируем нулевую и альтернативную гипотезу:\
Нулевую гипотезу сформулируем так, чтобы возможно было использовать знак равенства.\
H_0 - средние пользовательские рейтинги платформ Xbox One и PC одинаковые\
H_1 - средние пользовательские рейтинги платформ Xbox One и PC разные\
\
Для проверки гипотез будем использовать критерий уровень значимости — p-value, значение которого считаю в данном случае допустимым принять 5%.

Проверим гипотезы:

```python
xone_user_score = df_uscore[df_uscore['platform'] == 'XOne']['user_score']
pc_user_score = df_uscore[df_uscore['platform'] == 'PC']['user_score']
results = st.ttest_ind(
    xone_user_score,
    pc_user_score,
    equal_var = False)

alpha = .05
print(results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод:**\
Средние пользовательские рейтинги платформ Xbox One и PC одинаковые
